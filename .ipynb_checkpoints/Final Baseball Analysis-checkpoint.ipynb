## Intro
What statistics are most important in baseball? Which players get paid more? How much money will the average baseball player make in the future?There are countless questions worth examining, however there is one question that I will dive deeper into in this analysis.

Are there particular statistics in baseball  that correlate to some professionals earning more money than others?

In the following work, you will see analysis attempting to awnser this question. By taking a look at baseball players and their career statistics, I hope to determine what statistics are more likely to correlate to high earning athletes. To do this, we will import the collective stats of baseball players from 1871, and examine the career earnings of these players, and will run correlation calculations to examine which statistics are most likely to eplain higher earnings. 

```python
#Import Python Modules 
#Defining import function

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def import_stats(batting_csv,pitching_csv,players_csv):                             
    players= pd.read_csv(players_csv,\
                    usecols=['playerID','nameLast','nameGiven',\
                             'debut',"finalGame"]).copy()     # .copy to make sure we do not change original.
    batting= pd.read_csv(batting_csv).copy()
    batting.drop(['yearID','stint','teamID','lgID'],1,inplace=True)                                # Drop some less important stats.  
    pitching=pd.read_csv(pitching_csv).copy()
    pitching.drop(['yearID','IBB','GF','stint','teamID','lgID','SH','SF','GIDP'],1,inplace=True)   # Drop some less important stats. 
    pitching_avg=pitching[['ERA','BAOpp','playerID']].groupby('playerID',as_index=False).mean()    # average stats needed
    pitching.drop(['ERA','BAOpp'],1,inplace=True)                                                  #Get rid of these from DF
    pitching_sums=pitching.groupby('playerID',as_index=False).sum()
    pitching_totals=pitching_sums.merge(pitching_avg,on='playerID',how='inner')                    # Pitching stats summed or averaged
    batting_sums=batting.groupby('playerID',as_index=False).sum()                                  #All batting stats summed
    pitching_stats = players.merge(pitching_totals, on='playerID',how='inner')                     #Merge players with stats
    batting_stats =players.merge(batting_sums, on='playerID',how='inner')
    return (batting_stats, pitching_stats)
```

The function above was defined to give us a worakable Pandas Dataframe for our batting and pitching statistics. Since we are investigating the player's career stats with their career salary earnings, we will use the above function to import the statistics and create two seaperate Dataframes, one for pitchers and the other for batters. These players were seaperated because of the nature of the game of baseball, and the complete lack of overlap between bating and pitching statistics. What the function will do, is import the CSV files found from our source, and create dataframes of the players' personal info as well as the batting or pitching performance statistics. The stats we will compare for this analysis will be the following:

###### Batting:
Games, At Bats, Runs, Hits, Doubles, Triples, Home Runs, Runs Batted In, Stolen Bases, Caught Stealing, Walks, Strike Outs, Intentional Walks, Hit By Pitch, Sacrafice Hits, Sacrafice Flies, Grounded Into Double Plays
###### Pitching: 
Wins, Losses, Games, Games Statred, Complete Games, Shut Outs, Saves, Outs Pitched, Hits Allowed, Earned Runs Allowed, Home Runs Allowed, Walks, Strike Outs, Wild Pitches, Hit By Pitches Thrown, Balks, Batters Faced, Runs Given Up, Earned Run Average, Opponents Batting Average

We need to make sure to include not just the statistics of each player, but also the salary for each player. The code below imports all of the playing statistics, then merges the salary information for eachplayer into a new column. For this column, we will combine all salary earning for an individual player into one data point named "SumSalary" that will represent the career earnings for each player. Thus the final dataframes will have hall playing statistics as well as a career salary datapoint.

The output at the bottom is a reference of how many players are in each Dataframe. 

```python
#Importing the Data

batting_stats, pitching_stats = import_stats('/Users/ericgordo/Documents/Udacity/P2 Intro to Data Analysis/P2 Baseball Project/github_P2/Batting.csv',\
                                             '/Users/ericgordo/Documents/Udacity/P2 Intro to Data Analysis/P2 Baseball Project/github_P2/Pitching.csv',\
                                             '/Users/ericgordo/Documents/Udacity/P2 Intro to Data Analysis/P2 Baseball Project/github_P2/Master.csv')
   
salaries= pd.read_csv('/Users/ericgordo/Documents/Udacity/P2 Intro to Data Analysis/P2 Baseball Project/github_P2/Salaries.csv',\
                      usecols=['playerID','teamID','salary','yearID'])

sum_salary=salaries.groupby('playerID',as_index=False).sum()              # Turn Salaries into career Summed Salaries!!! 
sum_salary.rename(columns = {'salary': 'SumSalary'},inplace=True) 
sum_salary.drop(['yearID'],1,inplace=True)

pitchers_and_salaries =pitching_stats.merge(sum_salary, on='playerID',how='inner')
batters_and_salaries=batting_stats.merge(sum_salary,on='playerID',how='inner')

pitchers_and_salaries.dropna(inplace=True)
batters_and_salaries.dropna(inplace=True)

print "Pitchers:",len(pitchers_and_salaries)
print "Batters:",len(batters_and_salaries)
```

## Correlation Function and Exploration

Now that we have seen information about salaries and the data, we want to investigate which baseball stats have the highest relationship with players' career earnings. The function below will be used multiple times to compare the correlation between the total salary for a player, and each other stat. We will be using Pearson's Correlation for all correlation calculations. 

The function will only print the 4 highest correlated stats to salaries, and the 4 lowest correlated stats to salaries. 

```python
def find_corr(df,stat):
    corr= df.corr()[stat].sort_values(ascending=False)
    corr=corr.iloc[1:5],corr.tail(5)
    return corr
```

```python
print "Pitching Correlations"
print find_corr(pitchers_and_salaries,'SumSalary')
print ('\n')
print "Batting Correlations"
print find_corr(batters_and_salaries,'SumSalary')
```

After running this for the first time we can see a couple of things. 

For pitchers, it seems that the cumliative stats, like "Strikeouts", and "Wins" have the highest correlation to career earnings. The more strikeouts and wins a player accumilates over a career, the higher their career earnings will most likely be. The correlation coefficient for strikeouts has a value of r=.680, which is a fairly strong correlation. 

The negative correlation with "Earned Runed Average", and "Opponents Batting Average" makes snese, because these two stats are averages about how many runs and hits opponents have against them, which means that the lower these are, the better the pitcher (in theory). It is shocking however, that these two stats do not have a greater absolute value, because these stats are often considered very important in determining the ability of a pitcher. I would have expected these to be larger negative numbers, instead of the weak correlations of just r=-.078 and r=-.114, respectively.

For batters, "Home Runs" are the most telling stat of correlation to a players' career earnings, with "Runs Batted In" a close second. Also, "Sacrafice Hits" and "Caught Steeling" stats are the least predictive of career earnings, which I believe makes sense, because these stats are often not seen as crucial stats for judging a player's skills. This correlation coefficent calculated for the number of Home Runs a player hits over his career and his salary earnings was calculated to be r=.566.

Batting average is a simple calculation of taking the total number of hits a player has, and dividng that number by the total number of at bats. The code below will calculate batting averages and add it to our dataframe. 

```python
#Switch to Average Salaries! Should help with ERA?  Take another Comparison 
#Also Add BA, because that is my favorite known stat

avg_salary=salaries.groupby('playerID',as_index=False).mean()              # CREATE DF with AVERAGE SALARY, Drop Sum 
avg_salary.rename(columns = {'salary': 'AvgSalary'},inplace=True)


pitchers_avg=pitching_stats.merge(avg_salary, on='playerID',how='inner')
batters_avg=batting_stats.merge(avg_salary, on='playerID',how='inner')

batters_avg['BA']=batters_avg['H']/batters_avg['AB']                     #Add BA  to batting stats

print "Batters:",len(batters_avg)
print "Pitchers:",len(pitchers_avg)
```

Now having average salaries instead of summed salaries, we will run the correlation function again to see if we notice anything different.

```python
# Do correlations change compared to original calculations?

print "Pitching Correlations"
print find_corr(pitchers_avg,'AvgSalary')
print ('\n')
print "Batting Correlations"
print find_corr(batters_avg,'AvgSalary')

```

The correlations between every statistic and a player's average salary earnings are slightly weaker than was the case with career summed salary. The only exceptions to this were the two negatively correlated batting statistics, wich instead became stronger negative correlations. 

Interestingly, one stat for each batters and pitchers became more important in describing the correlation between performance and salary earnings. That is for pitchers, the number of players they "Hit with Pitches" became a stronger indicator of average salary earnings than Wins, and for Batters, "Strike Outs" became more predicitve then runs batted in. ## Batting Average Investigation
Finally, it is worth investigating why batting averages have such a low correlation with salary earnings. Even after cutting the players who have had less then 100 at bats it still is calculated as the weakest of all the correlation coefficients. Let's look at the scatterplots below to try to investigate. 

```python
color=sns.color_palette("Blues")
plt.scatter(batters_avg['BA'],batters_avg['AvgSalary'],c=color)

plt.xlabel('Batting Average')
plt.ylabel('Avg Salary')
plt.title('BA vs. Average Salary',fontsize=18)
```

In the above scatterplot, it is pretty clear that there is no clear correlation, and players are ranging anywhere from 0.05 to 0.35. As a fan of baseball I do know a couple of additional facts about batting averages. Most importantly though, is that it is very rare to see a professional batter hit less than .200 and for that same player to stay in the major leauge very long. The scatterplot above showed a significant number of players below .200, which could be from pitchers who have had many at bats, or players who were never considered very good. Thus in my last attempt to see if there was a reason for such the low correlation, I am additionally going to only look at players who had a batting average above .200. I will assume that this group of batters will now be everyday players, or at least batters who had a longer career.  Maybe with this final condition, we will be able to see a slight correlation between batting average and career salary earnings. 

```python
batters_avg= batters_avg[batters_avg['BA'] > 0.200]

plt.scatter(batters_avg['BA'],batters_avg['AvgSalary'],c=color)

plt.xlabel('Batting Average')
plt.ylabel('Average Salary')
plt.title('BA vs. Average Salary',fontsize=18)


print find_corr(batters_avg,'AvgSalary')
```

Now this scatterplot seems to have a slight correlation trend. This scatterplot visually seems to make a lot more sense. Also from looking at the calculation above, the correlation is now stronger, and is no longer one of the bottom 4 values in our calculations. 

This scatterplot I feel is a great final representation of players' batting averages compared to their expected career average salary earnings. While Batting Avearage is still not as predictive of a stat for a palyer's earnings as Home Runs or a few other statistics, it still is valued in the baseball community as a very important statistic.

If you are a baseball player, and you want to earn more in your career, it clearly seems that it is more likely to happen if you hit home runs as a batter, or throw strike outs as a pitcher. These stats seem to be the most correlated to higher career earnings than others like games played or even games won. While these stats are the most correlated to palyers' career salaries, there is still much room for further investigation. 

This simply summed career salaries, so it may be worth examining players's salaries on a normalized scale, or with some other factors into player salary. The statistics in this paper come from players all the way from 1871, but this paper did not account for ecconomic factors like inflation. There may be different results if the salary scales were normalized, or factored with the value of a salary for the time. Also this analysis did not take a look into the social changes in players' contracts over the long history of baseball. So while there was an investigation of how player stats are correlated to career earnings, futuer analysis may look into researching how contracts, salaries, and inflation have affected baseball players earnings.
